```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
sns.set_style("whitegrid")

df = pd.read_csv('../Day1_DataWrangling/data/cleaned_googleplaystore.csv')
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='Rating', bins=30, kde=True)
plt.title('Distribution of App Ratings on Google Play Store')
plt.xlabel('Rating')
plt.ylabel('Count of Apps')
plt.axvline(df['Rating'].mean(), color='red', linestyle='--', label=f'Mean: {df["Rating"].mean():.2f}')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(12, 8))
category_order = df['Category'].value_counts().index
sns.countplot(data=df, y='Category', order=category_order)
plt.title('App Categories by Count')
plt.xlabel('Number of Apps')
plt.ylabel('Category')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))

plot_df = df if df.shape[0] < 1000 else df.sample(n=1000, random_state=42)
sns.scatterplot(data=plot_df, x='Rating', y='Reviews', alpha=0.6) 
plt.title('App Rating vs. Number of Reviews')
plt.yscale('log') 
plt.show()
```

```python
# Plot 1: Free vs. Paid Count
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1) # 1 row, 2 columns, 1st plot
sns.countplot(data=df, x='Type')
plt.title('Count of Free vs. Paid Apps')

# Plot 2: Distribution of Prices for Paid Apps
plt.subplot(1, 2, 2) # 1 row, 2 columns, 2nd plot
paid_apps = df[df['Type'] == 'Paid']
sns.histplot(paid_apps['Price'], bins=100)
plt.title('Price Distribution of Paid Apps')
plt.xlabel('Price ($)')

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))

content_order = df['Content Rating'].value_counts().index
sns.countplot(data=df, y='Content Rating', order=content_order)
plt.title('Number of Apps by Content Rating')
plt.xlabel('Count')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
# Let's plot both distributions on the same chart to compare
sns.kdeplot(data=df, x='Rating', label='Original Rating', fill=True)
sns.kdeplot(data=df, x='Rating_normalized', label='Normalized Rating', fill=True)
plt.title('Comparison: Original vs. Normalized Rating')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(14, 8))
# Let's compare the top 5 categories
top_categories = df['Category'].value_counts().index[:5]
plot_df = df[df['Category'].isin(top_categories)]

sns.boxplot(data=plot_df, x='Category', y='Rating', hue='Type')
plt.title('App Rating Distribution: Free vs. Paid (Top 5 Categories)')
plt.xticks(rotation=45) # Rotate category names if they overlap
plt.tight_layout()
plt.show()
```
